# 3. DataViz

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### a. Choose a dataset, you can use Seaborn __[example datasets](https://seaborn.pydata.org/generated/seaborn.load_dataset.html)__.  
### Create a cheat sheet for yourself containing all plot types discussed in the lecture.
### Provide the following info:
- Plot type
- Use cases (categorical data, distribution, etc.)
- Example on the dataset

```python
tips = sns.load_dataset("tips")
tips
```

```python
tips.describe()
```

Histogram. Used for showing frequency/quantity correlation on the range of selected parameters. Can be applied to categorical and continous data.

```python
sns.histplot(tips['day'])
```

A kernel density estimate. Used to show distributions. Can be applied only to continous data.

```python
sns.kdeplot(data=tips, fill=True, x='tip', hue='smoker')
```

Empirical cumulative distributions. Can be used for continous data and categorical data. Shows % proportions of data.

```python
sns.ecdfplot(x='day', hue='smoker', data=tips)
```

Linear regression plot. Shows distribution of datapoints with two continous parameters and calculated linear approximation.

```python
sns.regplot(x='total_bill', y='tip', data=tips)
```

Categorical scatterplot or stripplot. Shows density and distribution of datapoints by categories.  
Technically can be used only with categorical/continous data. 
Should be used for conjunction of categorical and continous data.

```python
ax = sns.catplot(y="tip", x="size", data=tips, kind='strip')
```

Swarm plot. Shows distribution in categories with datapoints distinctly placed so that they do not overlap. Can be used for categorical data with continous data

```python
sns.catplot(y="tip", size=4, x="day", kind="swarm", data=tips)
```

Box plot. Shows statistic features of data such as median, quartiles, range in which belong most of the datapoints and outliers. 
Used on continous data for viewing it's distribution within categories.

```python
sns.catplot(x="time", y="total_bill", hue="sex", kind="box", data=tips)
```

Violin plot. Shows volumetric distribution of data. Used for continous data with categorical data

```python
sns.catplot(x="total_bill", y="day", hue="time", kind="violin", data=tips)
```

Combined plot. Combinations of plots can be used for presenting deeper insights into categorical and continous data 

```python
plot = sns.catplot(x="size", y="total_bill", inner=None, kind="violin", data=tips)
sns.stripplot(x="size", y="total_bill", size=3, color='k', data=tips, ax=plot.ax)
```
